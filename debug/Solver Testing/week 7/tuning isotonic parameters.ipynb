```julia
using Laplacians
```

```julia
include("../../../src/isotonicIPM.jl")
include("../../../src/samplingSolver.jl")
```

```julia
function generateInput(n,i)
    a = wtedChimera(n,i);
    b = zeros(Float64,n);
    b[1] = 1;
    b[n] = -1;
    v = cg(lap(a),b,verbose=false); # use any other solver
    v = v / (v[1] - v[n])

    p = sortperm(v);
    ap = a[p,p];
    # vp = v[p]
    vp = v[p] + rand(n) / sqrt(n);
    at = triu(ap);
    
    return at,vp
end
```

```julia
eps = 1e-5
a,v = generateInput(30000,5)
@time (x,acc,itercount) = isotonicIPMrelEps(a,v,eps,
(H -> samplingSolver(adj(H)[1], adj(H)[2], eps=0.5, sampConst=0.02, beta=1e2, verbose=true, tol=1e-1,maxits=1000)));
```
