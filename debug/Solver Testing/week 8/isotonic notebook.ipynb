```julia
function generateInput(n,i)
    a = wtedChimera(n,i);
    b = zeros(Float64,n);
    b[1] = 1;
    b[n] = -1;
    f = KMPLapSolver(a, verbose=true, tol=1e-1); # use any other solver
    v = f(b)
    v = v / (v[1] - v[n])

    p = sortperm(v);
    ap = symPermuteCSC(a, p);
    # vp = v[p]
    vp = v[p] + rand(n) / sqrt(n);
    at = triu(ap);
    
    return at,vp
end
```

```julia
@time a,v = generateInput(100000,23);
```

```julia
eps = 1e-5
@time (x,acc,itercount) = isotonicIPMrelEps(a,v,eps,
(H -> KMPSDDSolver(H, verbose=true, tol=1e-1)));
```
