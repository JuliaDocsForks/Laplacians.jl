```julia
tots = 0

for i in 1:150
    println("Working for ", i)
    
    n = 1000000;
    a = wtedChimera(n,i);
    b = rand(n); b = b - mean(b);
    
    tic()
    f = KMPLapSolver(a, verbose=true, tol=1e-1);
    f(b);
    tots += toc()
    
    println()
end

println("Total time = ", tots)
```
