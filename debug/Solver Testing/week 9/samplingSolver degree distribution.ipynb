```julia
using Laplacians
using PyPlot
```

```julia
include("../../../src/samplingSolver.jl")
```

```julia
n = 100000
a = chimera(n,1);
b = rand(n); b = b - mean(b);
```

### blowing it up by n^2

```julia
@time f,gOp,U,d,ord,cn,cntime = buildSolver(a, eps = 0.5, sampConst = 0.02, beta = 1e2, verbose=false);
checkError(gOp)
```

```julia
degs = []
for i in 1:n
    push!(degs, deg(U.data, i))
end
```

```julia
plot(1:n, degs)
```

### blowing it up by n^3

```julia
@time f,gOp,U,d,ord,cn,cntime = buildSolver(a, eps = 0.5, sampConst = 0.02, beta = (1e3)-50, verbose=false);
checkError(gOp)
```

```julia
degs = []
for i in 1:n
    push!(degs, deg(U.data, i))
end
```

```julia
plot(1:n, degs)
```
