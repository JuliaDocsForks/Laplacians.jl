See how degrees change throughout running the solver.

```julia
using Laplacians
using PyPlot
```

```julia
include("../../src/samplingSolver.jl")
```

```julia
n = 100000;
A = chimera(n, 1);
```

```julia
@time f,gOp,U,d = buildSolver(A, eps = 0.5, sampConst = 0.02);
```

```julia
checkError(gOp)
```

```julia
u = U.data;
```

```julia
degs = []

for i in 1:u.n
    push!(degs, nnz(u[:,i]))
end
```

```julia
plot(1:n, degs)
grid("on")
```

# average degree of chimera

```julia
chimdegs = []

for i in 1:A.n
    push!(chimdegs, nnz(A[:,i]))
end

maximum(chimdegs)
```
